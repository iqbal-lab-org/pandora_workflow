```python
import pandas as pd
```

```python
assemblies = pd.read_csv("assemblies_formatted.tsv", sep="\t")
assemblies
```

```python
reads = pd.read_csv("sra.tsv", sep="\t")
reads
```

```python
assemblies_and_reads = assemblies.merge(reads, how="inner", left_on="BioSample", right_on="sample_accession")
assemblies_and_reads
```

```python
from pathlib import Path
reads_paths = pd.read_csv("reads_paths.tsv")
reads_paths["run_accession"] = reads_paths["read_path"].apply(lambda read_name: Path(read_name).name.replace(".fastq", ""))
reads_paths
```

```python
assemblies_and_reads_paths = reads_paths.merge(assemblies_and_reads, how="inner", on="run_accession")
assemblies_and_reads_paths
```

```python
from pathlib import Path
assemblies_paths = pd.read_csv("assemblies_paths.tsv")
assemblies_paths["Assembly"] = assemblies_paths["assembly_path"].apply(lambda assembly_path: "GCA_" + Path(assembly_path).name.split("_")[1])
assemblies_paths
```

```python
assemblies_paths_and_reads_paths = assemblies_and_reads_paths.merge(assemblies_paths, how="inner", on="Assembly")
assemblies_paths_and_reads_paths
```

```python
assemblies_paths_and_reads_paths_simplified = assemblies_paths_and_reads_paths[["Assembly", "run_accession", "assembly_path", "read_path"]]
assemblies_paths_and_reads_paths_simplified.to_csv("assemblies_paths_and_reads_paths_simplified.csv")
assemblies_paths_and_reads_paths_simplified
```

```python
with open("script_to_build_input.sh", "w") as fout:
    print("set -eux", file=fout)
    data_dir="/hps/nobackup/research/zi/projects/pandora_analysis_pipeline/data_ecoli_92_way"
    for assembly, assembly_path, read_path in zip(assemblies_paths_and_reads_paths_simplified["Assembly"], assemblies_paths_and_reads_paths_simplified["assembly_path"], assemblies_paths_and_reads_paths_simplified["read_path"]):
        print(f"mkdir -p {data_dir}/{assembly}", file=fout)
        print(f"gunzip -c {assembly_path} > {data_dir}/{assembly}/{assembly}.ref.fa", file=fout)
        print(f"cp {read_path} {data_dir}/{assembly}/{assembly}.illumina.fastq", file=fout)
        print(f"gzip {data_dir}/{assembly}/{assembly}.illumina.fastq", file=fout)
```
