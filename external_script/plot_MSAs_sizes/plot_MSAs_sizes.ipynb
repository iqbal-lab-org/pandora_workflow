```python
import seaborn as sns
```

```python
msa_sizes = []
with open("msas_sizes.txt") as fin:
    for line in fin:
        msa_size = int(line.split()[0])
        msa_sizes.append(msa_size)
msa_sizes
```

```python
sns.histplot(msa_sizes, log_scale=(True, True))
```

```python
print(len(msa_sizes))
print(len([size for size in msa_sizes if size >= 400]))
print(len([size for size in msa_sizes if size >= 351]))
```
