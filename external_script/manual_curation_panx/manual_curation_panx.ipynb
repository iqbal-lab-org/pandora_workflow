```python
from glob import glob
```

```python
threshold_of_spaces_to_throw_sequence_out = 0.8

def get_headers_and_seqs(in_file):
    headers = []
    seqs = []
    seq=""
    for line in in_file:
        line = line.strip()
        if line.startswith(">"):
            previous_seq_is_empty = len(seq)==0
            if not previous_seq_is_empty:
                seqs.append(seq)
            headers.append(line)
            seq=""
        else:
            seq+=line
    previous_seq_is_empty = len(seq)==0
    if not previous_seq_is_empty:
        seqs.append(seq)
    assert len(headers) == len(seqs)
    return headers, seqs


def get_percentage_of_spaces_in_seq(seq):
    return seq.count("-") / len(seq)


def process(in_file, out_file, removed_seqs_file):
    headers, seqs = get_headers_and_seqs(in_file)
    for header, seq in zip(headers, seqs):
        percentage_of_spaces_in_seq = get_percentage_of_spaces_in_seq(seq)
        too_much_spaces = percentage_of_spaces_in_seq >= threshold_of_spaces_to_throw_sequence_out
        if not too_much_spaces:
            print(header, file=out_file)
            print(seq, file=out_file)
        else:
            print(f"Removing sequence ({percentage_of_spaces_in_seq} spaces): {header}")
            print(header, file=removed_seqs_file)
            print(seq, file=removed_seqs_file)
    

files = glob("genes_where_make_prg_took_10_GBs/*.fa")
for file in files:
    with open(file) as in_file, open(file+".curated.fa", "w") as out_file, open(file+".removed.fa", "w") as removed_seqs_file:
        process(in_file, out_file, removed_seqs_file)
```
