```python
import plotly.express as px
import pandas as pd
```

```python
# configs
clustalo_log_file = "clustalo_log.log"
make_prg_log_file = "make_prg_log.log"
```

```python
with open(clustalo_log_file) as fin:
    clustalo_logs = fin.readlines()
with open(make_prg_log_file) as fin:
    make_prg_logs = fin.readlines()
```

```python
def get_nbs_of_seqs_and_times_lists(log_lines, flag_line, nb_of_seqs_offset, time_offset):
    nbs_of_seqs = []
    times = []
    for index, line in enumerate(log_lines):
        if flag_line in line:
            nb_of_seqs = int(log_lines[index-nb_of_seqs_offset].strip().split()[-1])
            nbs_of_seqs.append(nb_of_seqs)
            time = float(log_lines[index-time_offset].strip().split()[-1])
            times.append(time)
    return nbs_of_seqs, times

clustalo_nbs_of_seqs, clustalo_times = get_nbs_of_seqs_and_times_lists(clustalo_logs, "Multiple sequence alignment finished.", 2, 1)
make_prg_nbs_of_seqs, make_prg_times = get_nbs_of_seqs_and_times_lists(make_prg_logs, "Finished building PRG.", 3, 1)
```

```python
df = pd.DataFrame(data={"clustalo_times_in_seconds": clustalo_times, "nb_of_sequences": clustalo_nbs_of_seqs})
px.scatter(df, x="clustalo_times_in_seconds", y="nb_of_sequences")
```

```python
df = pd.DataFrame(data={"clustalo_times_in_seconds": clustalo_times, "nb_of_sequences": clustalo_nbs_of_seqs})
px.scatter(df, x="clustalo_times_in_seconds", y="nb_of_sequences", log_x=True, log_y=True)
```

```python
df = pd.DataFrame(data={"make_prg_times_in_seconds": make_prg_times, "nb_of_sequences": make_prg_nbs_of_seqs})
px.scatter(df, x="make_prg_times_in_seconds", y="nb_of_sequences")
```

```python
df = pd.DataFrame(data={"make_prg_times_in_seconds": make_prg_times, "nb_of_sequences": make_prg_nbs_of_seqs})
px.scatter(df, x="make_prg_times_in_seconds", y="nb_of_sequences", log_x=True, log_y=True)
```
